```python
#Nous travaillons sur ce df global dans lequel il n'y a plus de valeur manquante ni de doublons pour le rendre exploitable
#dans le cadre de modélisation d'algorithme de machine learning

#A mon avis pour rendre cette base de donnée exloitable il faudrait encoder la variable qualitative 'regional indicator'
#Puis réaliser une normalisation ou une standardisation selon ce qui fonctionne le mieux. Nous effectuerons ces tests
```

```python
#Commençons par enlever les colonnes 'Country name' et 'year' que nous considérons inapproprié pour 
#le modele de machine learning
df_global2_nocountry=df_global2.drop(['Country name','year'], axis=1)
```

```python
df_global2_nocountry.head(3)
```

```python
#Avons de commencer les méthodes d'encodage, de normalisation ou encore de standardisation, commençons par séparer
#les variables explicatives de la variable cible

df_global2_y=df_global2_nocountry['Life Ladder']
df_global2_X=df_global2_nocountry.drop(['Life Ladder'],axis=1)
```

```python
#Puis séparont le jeu d'entrainement du jeu de text
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(df_global2_X,df_global2_y,test_size=0.20)
```

```python
X_train.head(5)
```

```python
#On peut maintenant encoder les colonnes 'Regional indicator' avec le OneHotEncoder car il ne s'agit pas de donnée ordinales
X_train=pd.get_dummies(X_train,columns=['Regional indicator'])
X_test=pd.get_dummies(X_test,columns=['Regional indicator'])
```

```python
#Ici j'essayais avec le OneHotEncoder mais comme sur l'examen je n'y arrive pas alors j'ai fais avec le get dummies
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder(drop='first',sparse=False)

cat=['Regional indicator']

X_train[cat]=ohe.fit_transform(X_train[cat])
X_test[cat]=ohe.transform(X_test[cat])

#Je dois essayer de comprendre le probleme du One Hot Encoder, j'y arrive avec get dummies mais impossible avvec le one hot *
#Demander à maxime ?
```

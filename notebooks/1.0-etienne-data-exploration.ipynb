```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 

import seaborn as sns

df = pd.read_csv('../data/world-happiness-report.csv')
df.head(20)
```

```python
print(df["Country name"].nunique())
```

```python
df.info()
```

```python
df_sorted = df.sort_values(by = 'Life Ladder', ascending = False)
df_sorted
```

```python
mean_per_year = df.groupby('year').mean()
mean_per_year
```

```python
# Normalisation des données pour que toutes les variables soient sur une échelle similaire
normalized_df = (mean_per_year - mean_per_year.min()) / (mean_per_year.max() - mean_per_year.min())
normalized_df
```

```python
plt.figure(figsize=(10, 6))

for column in normalized_df.columns:
    plt.plot(mean_per_year.index, normalized_df[column], label=column)

plt.legend()
plt.xlabel('Year')
plt.ylabel('Normalized Value')
plt.title('Evolution of Various Variables Over Time')
plt.grid(True)
plt.show()
```

```python
correlation_matrix = df.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
